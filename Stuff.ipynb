```python
### Date Fetching from the DataBase

date_list = []
for x in mycol.find({},{ "_id": 0, "actor": 1 }):
    y = x['actor']
    date_str = y["postedTime"][0:10]
#     date_time_obj = datetime.strptime(date_str, '%Y-%m-%d')
    print(date_str)
    date_list.append(date_str)
```

```python
date_list.sort()   ### Dates Sorting
```

```python
### Generate postedDates.txt file (Task-1 --> step-1  file)

with open('postedDates.txt', 'w', encoding='utf-8') as f:
        date = '\n'.join(date_list)
        f.write(date)
        f.close()
```

```python
#### Step-1 (Task -1)  complete
```

```python
### Now, We have to Count the Tweet for each day
```

```python
from collections import Counter
dates = date_list
c = Counter()
c.update(dates) 
```

```python
### Count The dates (Counted Number = each day Tweet Count)
count_dict = dict(c)
count_dict
```

```python
### Generate task1_output.txt file (Task-1 --> step-2  file)
with open('task1_output.txt', 'wt', encoding='utf-8') as f:
    for key in count_dict:
        f.write('"'+key+'"'+' '+str(count_dict[key])+'\n')
f.close()
```
